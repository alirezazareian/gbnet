```python
pred_sim = np.zeros((3, 51, 51))
for i in scene_graph_meta['idx_to_predicate']:
    for j in scene_graph_meta['idx_to_predicate']:
        if i == j:
            continue
        a = pred2synset[scene_graph_meta['idx_to_predicate'][i]]
        b = pred2synset[scene_graph_meta['idx_to_predicate'][j]]
        if a == '' or b == '':
            continue
        a = wn.synset(a)
        b = wn.synset(b)
        try:
            s1 = a.path_similarity(b)
        except:
            s1 = None
        try:
            s2 = a.lch_similarity(b)
        except:
            s2 = None
        try:
            s3 = a.wup_similarity(b)
        except:
            s3 = None
        if s1 == None: 
            s1 = 0.0
        if s2 == None: 
            s2 = 0.0
        else:
            s2 /= 3.6375861597263857
        if s3 == None: 
            s3 = 0.0
        pred_sim[:, int(i), int(j)] = [s1, s2, s3]
        
```
